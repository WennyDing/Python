```python
import sklearn
from sklearn.cluster import KMeans
import pandas as pd
%matplotlib inline
```

```python
data=pd.read_csv("d:\\pykmeans.csv")
```

```python
model=KMeans(n_clusters=5,max_iter=100)
```

```python
kmeansmodel=model.fit(data)
```

```python
kmeansmodel.cluster_centers_ #质心
```

```python
prdcluster=kmeansmodel.predict(data)
```

```python
prdcluster[0:100]#查看前100行对应的cluster_id
```
